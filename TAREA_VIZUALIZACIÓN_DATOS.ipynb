```python
# ============================================
# 1. CARGAR TUS DATOS
# ============================================
SIES = 'C:/Users/diego.perezp/Downloads/Matricula_2007_2025_WEB_15_07_2025.csv'
import pandas as pd

print(" Cargando archivo...")
df = pd.read_csv(SIES, sep=';', encoding='utf-8', low_memory=False)
print(f" Archivo cargado: {df.shape[0]:,} filas  {df.shape[1]} columnas\n")

# ============================================
# 2. PREPARAR DATOS PARA FILTROS
# ============================================
aos_disponibles = sorted(df['AO'].unique())
areas_disponibles = sorted(df['REA CARRERA GENRICA'].dropna().unique())
nivel_global_disponibles = sorted(df['NIVEL GLOBAL'].dropna().unique())
modalidad_disponibles = sorted(df['MODALIDAD'].dropna().unique())
niv3_disponibles = sorted(df['CLASIFICACIN INSTITUCIN NIVEL 3'].dropna().unique())
area_conocimiento_disponibles = sorted(df['REA DEL CONOCIMIENTO'].dropna().unique())

# ============================================
# 3. CREAR WIDGETS DE FILTRO (DESPLEGABLES)
# ============================================

filtro_aos = widgets.SelectMultiple(
    options=aos_disponibles,
    value=aos_disponibles,
    description='Aos:',
    layout=Layout(width='350px', height='180px')
)

filtro_areas = widgets.SelectMultiple(
    options=areas_disponibles,
    value=areas_disponibles,
    description='reas:',
    layout=Layout(width='350px', height='200px')
)

filtro_nivel_global = widgets.SelectMultiple(
    options=nivel_global_disponibles,
    value=tuple(nivel_global_disponibles),
    description='Nivel Global:',
    layout=Layout(width='350px', height='150px')
)

filtro_modalidad = widgets.SelectMultiple(
    options=modalidad_disponibles,
    value=tuple(modalidad_disponibles),
    description='Modalidad:',
    layout=Layout(width='350px', height='150px')
)

filtro_niv3 = widgets.SelectMultiple(
    options=niv3_disponibles,
    value=tuple(niv3_disponibles),
    description='Nivel 3:',
    layout=Layout(width='350px', height='200px')
)

filtro_area_conocimiento = widgets.SelectMultiple(
    options=area_conocimiento_disponibles,
    value=tuple(area_conocimiento_disponibles),
    description='rea Conocimiento:',
    layout=Layout(width='350px', height='200px')
)

# ============================================
# 4. BOTONES
# ============================================
btn_todos_aos = Button(description='Todos', button_style='info', layout=Layout(width='120px'))
btn_limpiar_aos = Button(description='Limpiar', button_style='warning', layout=Layout(width='120px'))

btn_todas_areas = Button(description='Todas', button_style='info', layout=Layout(width='120px'))
btn_limpiar_areas = Button(description='Limpiar', button_style='warning', layout=Layout(width='120px'))

btn_todos_nivel = Button(description='Todos', button_style='info', layout=Layout(width='120px'))
btn_limpiar_nivel = Button(description='Limpiar', button_style='warning', layout=Layout(width='120px'))

btn_todas_modalidad = Button(description='Todas', button_style='info', layout=Layout(width='120px'))
btn_limpiar_modalidad = Button(description='Limpiar', button_style='warning', layout=Layout(width='120px'))

btn_todos_niv3 = Button(description='Todos', button_style='info', layout=Layout(width='120px'))
btn_limpiar_niv3 = Button(description='Limpiar', button_style='warning', layout=Layout(width='120px'))

btn_todas_area_conocimiento = Button(description='Todas', button_style='info', layout=Layout(width='120px'))
btn_limpiar_area_conocimiento = Button(description='Limpiar', button_style='warning', layout=Layout(width='120px'))

btn_actualizar = Button(
    description=' ACTUALIZAR GRFICO',
    button_style='success',
    layout=Layout(width='300px', height='50px')
)

output_grafico = Output()
output_stats = Output()

# ============================================
# 5. FUNCIONES DE FILTROS
# ============================================
def seleccionar_todos_aos(b):
    filtro_aos.value = tuple(aos_disponibles)

def limpiar_aos(b):
    filtro_aos.value = ()

def seleccionar_todas_areas(b):
    filtro_areas.value = tuple(areas_disponibles)

def limpiar_areas(b):
    filtro_areas.value = ()

def seleccionar_todos_nivel(b):
    filtro_nivel_global.value = tuple(nivel_global_disponibles)

def limpiar_nivel(b):
    filtro_nivel_global.value = ()

def seleccionar_todas_modalidad(b):
    filtro_modalidad.value = tuple(modalidad_disponibles)

def limpiar_modalidad(b):
    filtro_modalidad.value = ()

def seleccionar_todos_niv3(b):
    filtro_niv3.value = tuple(niv3_disponibles)

def limpiar_niv3(b):
    filtro_niv3.value = ()

def seleccionar_todas_area_conocimiento(b):
    filtro_area_conocimiento.value = tuple(area_conocimiento_disponibles)

def limpiar_area_conocimiento(b):
    filtro_area_conocimiento.value = ()

btn_todos_aos.on_click(seleccionar_todos_aos)
btn_limpiar_aos.on_click(limpiar_aos)
btn_todas_areas.on_click(seleccionar_todas_areas)
btn_limpiar_areas.on_click(limpiar_areas)
btn_todos_nivel.on_click(seleccionar_todos_nivel)
btn_limpiar_nivel.on_click(limpiar_nivel)
btn_todas_modalidad.on_click(seleccionar_todas_modalidad)
btn_limpiar_modalidad.on_click(limpiar_modalidad)
btn_todos_niv3.on_click(seleccionar_todos_niv3)
btn_limpiar_niv3.on_click(limpiar_niv3)
btn_todas_area_conocimiento.on_click(seleccionar_todas_area_conocimiento)
btn_limpiar_area_conocimiento.on_click(limpiar_area_conocimiento)

# **Accordion con botones dentro**
accordion_filtros = widgets.Accordion(children=[
    VBox([filtro_aos, HBox([btn_todos_aos, btn_limpiar_aos])]),
    VBox([filtro_areas, HBox([btn_todas_areas, btn_limpiar_areas])]),
    VBox([filtro_nivel_global, HBox([btn_todos_nivel, btn_limpiar_nivel])]),
    VBox([filtro_modalidad, HBox([btn_todas_modalidad, btn_limpiar_modalidad])]),
    VBox([filtro_niv3, HBox([btn_todos_niv3, btn_limpiar_niv3])]),
    VBox([filtro_area_conocimiento, HBox([btn_todas_area_conocimiento, btn_limpiar_area_conocimiento])])
])
accordion_filtros.set_title(0, ' Aos')
accordion_filtros.set_title(1, ' rea Carrera Genrica')
accordion_filtros.set_title(2, ' Nivel Global')
accordion_filtros.set_title(3, ' Modalidad')
accordion_filtros.set_title(4, ' Nivel 3 Institucin')
accordion_filtros.set_title(5, ' rea del Conocimiento')

# ============================================
# 6. FUNCIN PARA ACTUALIZAR GRFICO
# ============================================
def actualizar_grafico(b):
    with output_grafico:
        clear_output(wait=True)
        
        aos_seleccionados = list(filtro_aos.value)
        areas_seleccionadas = list(filtro_areas.value)
        nivel_global_seleccionado = list(filtro_nivel_global.value)
        modalidad_seleccionada = list(filtro_modalidad.value)
        niv3_seleccionado = list(filtro_niv3.value)
        area_conocimiento_seleccionada = list(filtro_area_conocimiento.value)
        
        if not aos_seleccionados or not areas_seleccionadas:
            print(" Por favor, selecciona al menos un ao y un rea")
            return
        
        df_filtrado = df[
            (df['AO'].isin(aos_seleccionados)) &
            (df['REA CARRERA GENRICA'].isin(areas_seleccionadas)) &
            (df['NIVEL GLOBAL'].isin(nivel_global_seleccionado)) &
            (df['MODALIDAD'].isin(modalidad_seleccionada)) &
            (df['CLASIFICACIN INSTITUCIN NIVEL 3'].isin(niv3_seleccionado)) &
            (df['REA DEL CONOCIMIENTO'].isin(area_conocimiento_seleccionada))
        ].copy()
        
        if df_filtrado.empty:
            print(" No hay datos para los filtros seleccionados")
            return
        
        df_agrupado = df_filtrado.groupby(['AO', 'REA CARRERA GENRICA'])['TOTAL MATRCULA'].sum().reset_index()
        
        fig = px.line(
            df_agrupado,
            x='AO',
            y='TOTAL MATRCULA',
            color='REA CARRERA GENRICA',
            title='Evolucin de Matrcula Total por rea de Carrera Genrica',
            markers=True
        )
        
        fig.update_layout(
            hovermode='x unified',
            legend=dict(orientation='v', y=1, x=1.02),
            height=600,
            template='plotly_white'
        )
        fig.update_xaxes(dtick=1)
        fig.update_yaxes(tickformat=',')
        fig.show()
    
    with output_stats:
        clear_output(wait=True)
        
        print("="*80)
        print(" ESTADSTICAS DE LOS DATOS FILTRADOS")
        print("="*80)
        print(f" Aos seleccionados: {min(aos_seleccionados)} - {max(aos_seleccionados)}")
        print(f" reas seleccionadas: {len(areas_seleccionadas)}")
        print(f" Nivel Global: {nivel_global_seleccionado}")
        print(f" Modalidad: {modalidad_seleccionada}")
        print(f" Nivel 3: {niv3_seleccionado}")
        print(f" rea Conocimiento: {area_conocimiento_seleccionada}")
        print(f" Total registros: {len(df_filtrado):,}")
        print(f" Matrcula Total: {df_filtrado['TOTAL MATRCULA'].sum():,}")
        print("="*80)

        tabla = df_filtrado.groupby('REA CARRERA GENRICA')['TOTAL MATRCULA'].sum().sort_values(ascending=False)
        display(tabla)

btn_actualizar.on_click(actualizar_grafico)

# ============================================
# 7. DISEO DEL DASHBOARD
# ============================================
titulo = widgets.HTML(
    value="<h1 style='text-align:center; color:#2c3e50; background:#ecf0f1; padding:20px; border-radius:10px;'> Dashboard de Matrcula por rea y Ao</h1>"
)

panel_filtros = VBox([
    accordion_filtros
], layout=Layout(padding='10px'))

panel_control = VBox([
    btn_actualizar,
    widgets.HTML(value="<p style='text-align:center; color:#7f8c8d;'><i>Selecciona filtros y presiona actualizar</i></p>")
], layout=Layout(align_items='center', padding='10px'))

dashboard = VBox([
    titulo,
    panel_filtros,
    panel_control,
    output_stats,
    output_grafico
], layout=Layout(border='3px solid #2c3e50', padding='20px', border_radius='15px'))

# ============================================
# 8. MOSTRAR DASHBOARD
# ============================================
display(dashboard)

# ============================================
# 9. NUEVO GRFICO: MATRCULA POR REGIN, AO
# ============================================


from IPython.display import clear_output
import plotly.express as px

# --- 9.1 Preparamos listas para filtros ---
niv1_opts = sorted(df['CLASIFICACIN INSTITUCIN NIVEL 1'].dropna().unique())
niv3_opts = sorted(df['CLASIFICACIN INSTITUCIN NIVEL 3'].dropna().unique())
area_opts = sorted(df['REA CARRERA GENRICA'].dropna().unique())
inst_opts = sorted(df['NOMBRE INSTITUCIN'].dropna().unique())
nivel_global_opts = sorted(df['NIVEL GLOBAL'].dropna().unique())
modalidad_opts = sorted(df['MODALIDAD'].dropna().unique())

# --------- 9.2 WIDGETS --------------
f_n1 = widgets.SelectMultiple(
    options=niv1_opts,
    value=niv1_opts,
    description="Nivel 1:",
    layout=Layout(width='350px', height='200px')
)

f_n3 = widgets.SelectMultiple(
    options=niv3_opts,
    value=niv3_opts,
    description="Nivel 3:",
    layout=Layout(width='350px', height='200px')
)

f_area2 = widgets.SelectMultiple(
    options=area_opts,
    value=area_opts,
    description="reas:",
    layout=Layout(width='350px', height='200px')
)

f_inst = widgets.SelectMultiple(
    options=inst_opts,
    value=inst_opts,
    description="Institucin:",
    layout=Layout(width='350px', height='200px')
)

f_nivel_global = widgets.SelectMultiple(
    options=nivel_global_opts,
    value=tuple(nivel_global_opts),
    description="Nivel Global:",
    layout=Layout(width='350px', height='150px')
)

f_modalidad = widgets.SelectMultiple(
    options=modalidad_opts,
    value=tuple(modalidad_opts),
    description="Modalidad:",
    layout=Layout(width='350px', height='150px')
)

# -------- BOTONES seleccin/limpieza --------
btn_all_n1 = Button(description='Todos N1', button_style='info', layout=Layout(width='120px'))
btn_clr_n1 = Button(description='Limpiar N1', button_style='warning', layout=Layout(width='120px'))

btn_all_n3 = Button(description='Todos N3', button_style='info', layout=Layout(width='120px'))
btn_clr_n3 = Button(description='Limpiar N3', button_style='warning', layout=Layout(width='120px'))

btn_all_area2 = Button(description='Todas reas', button_style='info', layout=Layout(width='120px'))
btn_clr_area2 = Button(description='Limpiar reas', button_style='warning', layout=Layout(width='120px'))

btn_all_inst = Button(description='Todas Inst.', button_style='info', layout=Layout(width='120px'))
btn_clr_inst = Button(description='Limpiar Inst.', button_style='warning', layout=Layout(width='120px'))

btn_all_nivel_global = Button(description='Todos', button_style='info', layout=Layout(width='120px'))
btn_clr_nivel_global = Button(description='Limpiar', button_style='warning', layout=Layout(width='120px'))

btn_all_modalidad = Button(description='Todas', button_style='info', layout=Layout(width='120px'))
btn_clr_modalidad = Button(description='Limpiar', button_style='warning', layout=Layout(width='120px'))

# ---- Acciones de los botones ----
btn_all_n1.on_click(lambda b: setattr(f_n1, 'value', tuple(niv1_opts)))
btn_clr_n1.on_click(lambda b: setattr(f_n1, 'value', ()))

btn_all_n3.on_click(lambda b: setattr(f_n3, 'value', tuple(niv3_opts)))
btn_clr_n3.on_click(lambda b: setattr(f_n3, 'value', ()))

btn_all_area2.on_click(lambda b: setattr(f_area2, 'value', tuple(area_opts)))
btn_clr_area2.on_click(lambda b: setattr(f_area2, 'value', ()))

btn_all_inst.on_click(lambda b: setattr(f_inst, 'value', tuple(inst_opts)))
btn_clr_inst.on_click(lambda b: setattr(f_inst, 'value', ()))

btn_all_nivel_global.on_click(lambda b: setattr(f_nivel_global, 'value', tuple(nivel_global_opts)))
btn_clr_nivel_global.on_click(lambda b: setattr(f_nivel_global, 'value', ()))

btn_all_modalidad.on_click(lambda b: setattr(f_modalidad, 'value', tuple(modalidad_opts)))
btn_clr_modalidad.on_click(lambda b: setattr(f_modalidad, 'value', ()))

accordion2 = widgets.Accordion(children=[
    VBox([f_n1, HBox([btn_all_n1, btn_clr_n1])]),
    VBox([f_n3, HBox([btn_all_n3, btn_clr_n3])]),
    VBox([f_area2, HBox([btn_all_area2, btn_clr_area2])]),
    VBox([f_inst, HBox([btn_all_inst, btn_clr_inst])]),
    VBox([f_nivel_global, HBox([btn_all_nivel_global, btn_clr_nivel_global])]),
    VBox([f_modalidad, HBox([btn_all_modalidad, btn_clr_modalidad])])
])

accordion2.set_title(0, " Nivel 1")
accordion2.set_title(1, " Nivel 3")
accordion2.set_title(2, " rea Genrica")
accordion2.set_title(3, " Institucin")
accordion2.set_title(4, " Nivel Global")
accordion2.set_title(5, " Modalidad")

btn_update2 = Button(
    description=" ACTUALIZAR GRFICO 2",
    button_style="info",
    layout=Layout(width='300px', height='45px')
)

out_graph2 = Output()
out_stats2 = Output()

# ============================================
# 9.3 Lgica del grfico 2
# ============================================
def actualizar_grafico2(b):
    with out_graph2:
        clear_output(wait=True)
        
        temp = df.copy()
        
        # Aplicar filtros reales
        if f_n1.value:
            temp = temp[temp['CLASIFICACIN INSTITUCIN NIVEL 1'].isin(f_n1.value)]
        if f_n3.value:
            temp = temp[temp['CLASIFICACIN INSTITUCIN NIVEL 3'].isin(f_n3.value)]
        if f_area2.value:
            temp = temp[temp['REA CARRERA GENRICA'].isin(f_area2.value)]
        if f_inst.value:
            temp = temp[temp['NOMBRE INSTITUCIN'].isin(f_inst.value)]
        if f_nivel_global.value:
            temp = temp[temp['NIVEL GLOBAL'].isin(f_nivel_global.value)]
        if f_modalidad.value:
            temp = temp[temp['MODALIDAD'].isin(f_modalidad.value)]
        
        if temp.empty:
            print(" No hay datos para los filtros seleccionados.")
            return
        
        # Crear datos agregados para el grfico
        df_g = temp.groupby(['REGIN', 'AO'])['TOTAL MATRCULA'].sum().reset_index()
        
        # Crear texto detallado para hover (tooltip)
        hover_text = []
        for idx, row in df_g.iterrows():
            region = row['REGIN']
            anio = row['AO']
            
            # Filtrar datos especficos de esa regin y ao
            detalle = temp[(temp['REGIN'] == region) & (temp['AO'] == anio)]
            
            # Construir el texto del tooltip
            texto = f"<b>{region} - {anio}</b><br>"
            texto += f"<b>Total Matrcula: {row['TOTAL MATRCULA']:,}</b><br><br>"
            
            # Desglose por Nivel 1
            if len(detalle['CLASIFICACIN INSTITUCIN NIVEL 1'].unique()) > 0:
                texto += "<b>Por Nivel 1:</b><br>"
                nivel1_agg = detalle.groupby('CLASIFICACIN INSTITUCIN NIVEL 1')['TOTAL MATRCULA'].sum().sort_values(ascending=False)
                for nivel, mat in nivel1_agg.head(5).items():
                    texto += f"   {nivel}: {mat:,}<br>"
                if len(nivel1_agg) > 5:
                    texto += f"   ... y {len(nivel1_agg)-5} ms<br>"
                texto += "<br>"
            
            # Desglose por rea
            if len(detalle['REA CARRERA GENRICA'].unique()) > 0:
                texto += "<b>Por rea:</b><br>"
                area_agg = detalle.groupby('REA CARRERA GENRICA')['TOTAL MATRCULA'].sum().sort_values(ascending=False)
                for area, mat in area_agg.head(5).items():
                    texto += f"   {area}: {mat:,}<br>"
                if len(area_agg) > 5:
                    texto += f"   ... y {len(area_agg)-5} ms<br>"
                texto += "<br>"
            
            # Desglose por Institucin (top 5)
            if len(detalle['NOMBRE INSTITUCIN'].unique()) > 0:
                texto += "<b>Top Instituciones:</b><br>"
                inst_agg = detalle.groupby('NOMBRE INSTITUCIN')['TOTAL MATRCULA'].sum().sort_values(ascending=False)
                for inst, mat in inst_agg.head(5).items():
                    texto += f"   {inst}: {mat:,}<br>"
                if len(inst_agg) > 5:
                    texto += f"   ... y {len(inst_agg)-5} ms<br>"
            
            hover_text.append(texto)
        
        df_g['hover_text'] = hover_text
        
        # Crear grfico con hover personalizado
        fig2 = px.bar(
            df_g,
            x='REGIN',
            y='TOTAL MATRCULA',
            color='AO',
            barmode='group',
            title="Matrcula por Regin y Ao",
            custom_data=['hover_text']
        )
        
        # Configurar el hover template
        fig2.update_traces(
            hovertemplate='%{customdata[0]}<extra></extra>'
        )
        
        fig2.update_layout(
            template='plotly_white', 
            height=600,
            hovermode='closest'
        )
        fig2.update_yaxes(tickformat=',')
        fig2.show()
    
    with out_stats2:
        clear_output(wait=True)
        print("=" * 80)
        print(" ESTADSTICAS DEL GRFICO 2")
        print("=" * 80)
        print(f" Nivel 1 seleccionado: {list(f_n1.value)}")
        print(f" Nivel 3 seleccionado: {list(f_n3.value)}")
        print(f" rea(s) seleccionada(s): {list(f_area2.value)}")
        print(f" Institucin(es) seleccionada(s): {list(f_inst.value)}")
        print(f" Nivel Global: {list(f_nivel_global.value)}")
        print(f" Modalidad: {list(f_modalidad.value)}")
        print(f"\n Total registros filtrados: {len(temp):,}")
        print(f" Total Matrcula: {temp['TOTAL MATRCULA'].sum():,}")
        print("=" * 80)

btn_update2.on_click(actualizar_grafico2)

# ============================================
# 9.4 Instrucciones visuales
# ============================================
titulo2 = widgets.HTML(
    value="<h2 style='text-align:center; color:#2c3e50;'> Grfico 2: Matrcula por Regin y Ao</h2>"
)

instrucciones = widgets.HTML(value="""
<div style='background-color:#E8F8F5; padding:15px; border-radius:10px; border:2px solid #1ABC9C; margin:10px 0;'>
    <p style='margin:5px 0; font-size:14px;'><strong> Cmo usar los filtros:</strong></p>
    <ul style='margin:5px 0; font-size:13px;'>
        <li>Mantn presionado <strong>Ctrl</strong> (Windows/Linux) o <strong>Cmd</strong> (Mac) mientras haces clic para seleccionar mltiples opciones</li>
        <li>Usa los botones <strong>"Todos"</strong> para seleccionar todas las opciones disponibles</li>
        <li>Usa los botones <strong>"Limpiar"</strong> para deseleccionar todo</li>
        <li>Haz clic en <strong>" ACTUALIZAR GRFICO 2"</strong> para generar el grfico con los filtros seleccionados</li>
    </ul>
</div>
""")

# ============================================
# 9.5 Mostrar dashboard 2
# ============================================
dashboard2 = VBox([
    titulo2,
    instrucciones,
    accordion2,
    btn_update2,
    out_stats2,
    out_graph2
], layout=Layout(border='3px solid #2471A3', padding='20px', border_radius='15px'))

display(dashboard2)

print(" Dashboard listo. Haz clic en 'ACTUALIZAR GRFICO 2' para comenzar.")
print(" TIP: Usa Ctrl/Cmd + clic para seleccionar mltiples opciones.")

# ============================================
# GRFICO 3: DISPERSIN DE MATRCULA POR REA - COMPARACIN ENTRE INSTITUCIONES
# ============================================

import ipywidgets as widgets
from ipywidgets import HBox, VBox, Layout, Button, Output
from IPython.display import clear_output
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# --- Preparamos listas para filtros ---
niv1_opts_g3 = sorted(df['CLASIFICACIN INSTITUCIN NIVEL 1'].dropna().unique())
niv2_opts_g3 = sorted(df['CLASIFICACIN INSTITUCIN NIVEL 2'].dropna().unique())
niv3_opts_g3 = sorted(df['CLASIFICACIN INSTITUCIN NIVEL 3'].dropna().unique())
anio_opts_g3 = sorted(df['AO'].dropna().unique())
region_opts_g3 = sorted(df['REGIN'].dropna().unique())
inst_opts_g3 = sorted(df['NOMBRE INSTITUCIN'].dropna().unique())
nivel_global_opts_g3 = sorted(df['NIVEL GLOBAL'].dropna().unique())
jornada_opts_g3 = sorted(df['JORNADA'].dropna().unique())

# --------- WIDGETS --------------
f_niv1_g3 = widgets.SelectMultiple(
    options=niv1_opts_g3,
    value=niv1_opts_g3,
    description="Nivel 1:",
    layout=Layout(width='350px', height='150px')
)

f_niv2_g3 = widgets.SelectMultiple(
    options=niv2_opts_g3,
    value=niv2_opts_g3,
    description="Nivel 2:",
    layout=Layout(width='350px', height='150px')
)

f_niv3_g3 = widgets.SelectMultiple(
    options=niv3_opts_g3,
    value=niv3_opts_g3,
    description="Nivel 3:",
    layout=Layout(width='350px', height='150px')
)

# AO - Seleccin nica (Dropdown)
f_anio_g3 = widgets.Dropdown(
    options=anio_opts_g3,
    value=anio_opts_g3[-1] if len(anio_opts_g3) > 0 else None,
    description="Ao:",
    layout=Layout(width='350px')
)

f_region_g3 = widgets.SelectMultiple(
    options=region_opts_g3,
    value=region_opts_g3,
    description="Regin:",
    layout=Layout(width='350px', height='150px')
)

# INSTITUCIN - Seleccin mltiple (USS siempre incluida)
f_inst_g3 = widgets.SelectMultiple(
    options=inst_opts_g3,
    value=tuple(['UNIVERSIDAD SAN SEBASTIAN'] if 'UNIVERSIDAD SAN SEBASTIAN' in inst_opts_g3 else []),
    description="Institucin:",
    layout=Layout(width='350px', height='180px')
)

# NIVEL GLOBAL - Seleccin mltiple
f_nivel_global_g3 = widgets.SelectMultiple(
    options=nivel_global_opts_g3,
    value=tuple(nivel_global_opts_g3),
    description="Nivel Global:",
    layout=Layout(width='350px', height='150px')
)

# JORNADA - Seleccin mltiple
f_jornada_g3 = widgets.SelectMultiple(
    options=jornada_opts_g3,
    value=tuple(jornada_opts_g3),
    description="Jornada:",
    layout=Layout(width='350px', height='150px')
)

# -------- BOTONES seleccin/limpieza --------
btn_all_niv1_g3 = Button(description='Todos', button_style='info', layout=Layout(width='100px'))
btn_clr_niv1_g3 = Button(description='Limpiar', button_style='warning', layout=Layout(width='100px'))

btn_all_niv2_g3 = Button(description='Todos', button_style='info', layout=Layout(width='100px'))
btn_clr_niv2_g3 = Button(description='Limpiar', button_style='warning', layout=Layout(width='100px'))

btn_all_niv3_g3 = Button(description='Todos', button_style='info', layout=Layout(width='100px'))
btn_clr_niv3_g3 = Button(description='Limpiar', button_style='warning', layout=Layout(width='100px'))

btn_all_region_g3 = Button(description='Todos', button_style='info', layout=Layout(width='100px'))
btn_clr_region_g3 = Button(description='Limpiar', button_style='warning', layout=Layout(width='100px'))

btn_all_inst_g3 = Button(description='Todas', button_style='info', layout=Layout(width='100px'))
btn_clr_inst_g3 = Button(description='Limpiar', button_style='warning', layout=Layout(width='100px'))

btn_all_nivel_global_g3 = Button(description='Todos', button_style='info', layout=Layout(width='100px'))
btn_clr_nivel_global_g3 = Button(description='Limpiar', button_style='warning', layout=Layout(width='100px'))

btn_all_jornada_g3 = Button(description='Todas', button_style='info', layout=Layout(width='100px'))
btn_clr_jornada_g3 = Button(description='Limpiar', button_style='warning', layout=Layout(width='100px'))

# Acciones de los botones
btn_all_niv1_g3.on_click(lambda b: setattr(f_niv1_g3, 'value', tuple(niv1_opts_g3)))
btn_clr_niv1_g3.on_click(lambda b: setattr(f_niv1_g3, 'value', ()))

btn_all_niv2_g3.on_click(lambda b: setattr(f_niv2_g3, 'value', tuple(niv2_opts_g3)))
btn_clr_niv2_g3.on_click(lambda b: setattr(f_niv2_g3, 'value', ()))

btn_all_niv3_g3.on_click(lambda b: setattr(f_niv3_g3, 'value', tuple(niv3_opts_g3)))
btn_clr_niv3_g3.on_click(lambda b: setattr(f_niv3_g3, 'value', ()))

btn_all_region_g3.on_click(lambda b: setattr(f_region_g3, 'value', tuple(region_opts_g3)))
btn_clr_region_g3.on_click(lambda b: setattr(f_region_g3, 'value', ()))

btn_all_inst_g3.on_click(lambda b: setattr(f_inst_g3, 'value', tuple(inst_opts_g3)))
btn_clr_inst_g3.on_click(lambda b: setattr(f_inst_g3, 'value', ()))

btn_all_nivel_global_g3.on_click(lambda b: setattr(f_nivel_global_g3, 'value', tuple(nivel_global_opts_g3)))
btn_clr_nivel_global_g3.on_click(lambda b: setattr(f_nivel_global_g3, 'value', ()))

btn_all_jornada_g3.on_click(lambda b: setattr(f_jornada_g3, 'value', tuple(jornada_opts_g3)))
btn_clr_jornada_g3.on_click(lambda b: setattr(f_jornada_g3, 'value', ()))

# Accordion
accordion3 = widgets.Accordion(children=[
    VBox([f_niv1_g3, HBox([btn_all_niv1_g3, btn_clr_niv1_g3])]),
    VBox([f_niv2_g3, HBox([btn_all_niv2_g3, btn_clr_niv2_g3])]),
    VBox([f_niv3_g3, HBox([btn_all_niv3_g3, btn_clr_niv3_g3])]),
    VBox([f_anio_g3]),
    VBox([f_region_g3, HBox([btn_all_region_g3, btn_clr_region_g3])]),
    VBox([f_inst_g3, HBox([btn_all_inst_g3, btn_clr_inst_g3])]),
    VBox([f_nivel_global_g3, HBox([btn_all_nivel_global_g3, btn_clr_nivel_global_g3])]),
    VBox([f_jornada_g3, HBox([btn_all_jornada_g3, btn_clr_jornada_g3])])
])

accordion3.set_title(0, " Nivel 1")
accordion3.set_title(1, " Nivel 2")
accordion3.set_title(2, " Nivel 3")
accordion3.set_title(3, " Ao")
accordion3.set_title(4, " Regin")
accordion3.set_title(5, " Instituciones a Comparar")
accordion3.set_title(6, " Nivel Global")
accordion3.set_title(7, " Jornada")

btn_update3 = Button(
    description=" GENERAR GRFICO 3",
    button_style="success",
    layout=Layout(width='300px', height='45px')
)

out_graph3 = Output()
out_stats3 = Output()

# ============================================
# Lgica del grfico 3
# ============================================
def actualizar_grafico3(b):
    with out_graph3:
        clear_output(wait=True)
        
        temp = df.copy()
        
        # Aplicar filtros base
        if f_niv1_g3.value:
            temp = temp[temp['CLASIFICACIN INSTITUCIN NIVEL 1'].isin(f_niv1_g3.value)]
        if f_niv2_g3.value:
            temp = temp[temp['CLASIFICACIN INSTITUCIN NIVEL 2'].isin(f_niv2_g3.value)]
        if f_niv3_g3.value:
            temp = temp[temp['CLASIFICACIN INSTITUCIN NIVEL 3'].isin(f_niv3_g3.value)]
        if f_anio_g3.value:
            temp = temp[temp['AO'] == f_anio_g3.value]
        if f_region_g3.value:
            temp = temp[temp['REGIN'].isin(f_region_g3.value)]
        if f_nivel_global_g3.value:
            temp = temp[temp['NIVEL GLOBAL'].isin(f_nivel_global_g3.value)]
        if f_jornada_g3.value:
            temp = temp[temp['JORNADA'].isin(f_jornada_g3.value)]
        
        # ASEGURAR QUE USS SIEMPRE EST INCLUIDA
        uss_name = 'UNIVERSIDAD SAN SEBASTIAN'
        instituciones_seleccionadas = list(f_inst_g3.value) if f_inst_g3.value else []
        
        if uss_name not in instituciones_seleccionadas and uss_name in inst_opts_g3:
            instituciones_seleccionadas.append(uss_name)
        
        if not instituciones_seleccionadas:
            print(" Debes seleccionar al menos una institucin.")
            return
        
        # Filtrar por instituciones seleccionadas
        temp = temp[temp['NOMBRE INSTITUCIN'].isin(instituciones_seleccionadas)]
        
        if temp.empty:
            print(" No hay datos para los filtros seleccionados.")
            return
        
        # Agrupar por REA y NOMBRE INSTITUCIN
        df_plot = temp.groupby(['REA CARRERA GENRICA', 'NOMBRE INSTITUCIN'])['TOTAL MATRCULA'].sum().reset_index()
        df_plot = df_plot[df_plot['TOTAL MATRCULA'] > 0]
        
        if df_plot.empty:
            print(" No hay datos con matrcula > 0.")
            return
        
        # Ordenar reas por matrcula total
        area_order = df_plot.groupby('REA CARRERA GENRICA')['TOTAL MATRCULA'].sum().sort_values(ascending=True)
        areas_ordenadas = area_order.index.tolist()
        
        # Funcin para formatear etiquetas largas en dos lneas
        def formatear_etiqueta(texto, max_chars=30):
            if len(texto) <= max_chars:
                return texto
            # Buscar un espacio cerca de la mitad para el salto de lnea
            mitad = len(texto) // 2
            # Buscar el espacio ms cercano a la mitad
            espacio_antes = texto.rfind(' ', 0, mitad + 10)
            espacio_despues = texto.find(' ', mitad)
            
            if espacio_antes > mitad - 10:
                return texto[:espacio_antes] + '<br>' + texto[espacio_antes+1:]
            elif espacio_despues > 0:
                return texto[:espacio_despues] + '<br>' + texto[espacio_despues+1:]
            else:
                # Si no hay espacios, truncar
                return texto[:max_chars-3] + '...'
        
        # Crear versiones formateadas de las reas para el eje Y
        areas_formateadas = [formatear_etiqueta(area) for area in areas_ordenadas]
        area_map = {area_orig: area_fmt for area_orig, area_fmt in zip(areas_ordenadas, areas_formateadas)}
        
        # Crear figura de Plotly
        fig = go.Figure()
        
        # Colores para instituciones
        instituciones_unicas = df_plot['NOMBRE INSTITUCIN'].unique()
        n_inst = len(instituciones_unicas)
        
        # Paleta de colores
        colores_plotly = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', 
                          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
        colores = {}
        color_idx = 0
        
        for inst in instituciones_unicas:
            if inst == uss_name:
                colores[inst] = '#FF6B6B'
            else:
                colores[inst] = colores_plotly[color_idx % len(colores_plotly)]
                color_idx += 1
        
        # Agregar cada institucin como una trace
        for inst in instituciones_unicas:
            df_inst = df_plot[df_plot['NOMBRE INSTITUCIN'] == inst]
            
            # Construir hover text detallado para cada punto
            hover_texts = []
            for idx, row in df_inst.iterrows():
                area = row['REA CARRERA GENRICA']
                matricula = row['TOTAL MATRCULA']
                
                # Obtener detalles del rea e institucin
                detalle = temp[
                    (temp['REA CARRERA GENRICA'] == area) &
                    (temp['NOMBRE INSTITUCIN'] == inst)
                ]
                
                texto = f"<b>{inst}</b><br>"
                texto += f"<b>{area}</b><br>"
                texto += "<br>"
                texto += f"<b>Matrcula Total: {matricula:,.0f}</b><br><br>"
                
                if not detalle.empty:
                    # Desglose por Nivel 1
                    nivel1_data = detalle.groupby('CLASIFICACIN INSTITUCIN NIVEL 1')['TOTAL MATRCULA'].sum()
                    if len(nivel1_data) > 0:
                        texto += "<b>Por Nivel:</b><br>"
                        for nivel, mat in nivel1_data.items():
                            texto += f"   {nivel}: {mat:,.0f}<br>"
                        texto += "<br>"
                    
                    # Desglose por Jornada
                    jornada_data = detalle.groupby('JORNADA')['TOTAL MATRCULA'].sum()
                    if len(jornada_data) > 0:
                        texto += "<b>Por Jornada:</b><br>"
                        for jornada, mat in jornada_data.items():
                            texto += f"   {jornada}: {mat:,.0f}<br>"
                        texto += "<br>"
                    
                    # Desglose por Regin (top 3)
                    region_data = detalle.groupby('REGIN')['TOTAL MATRCULA'].sum().sort_values(ascending=False)
                    if len(region_data) > 0:
                        texto += "<b>Top Regiones:</b><br>"
                        for region, mat in region_data.head(3).items():
                            texto += f"   {region}: {mat:,.0f}<br>"
                        if len(region_data) > 3:
                            texto += f"   ... y {len(region_data)-3} ms<br>"
                
                hover_texts.append(texto)
            
            # Configurar marcador
            if inst == uss_name:
                marker_config = dict(
                    size=14,
                    color=colores[inst],
                    line=dict(color='white', width=2)
                )
            else:
                marker_config = dict(
                    size=10,
                    color=colores[inst]
                )
            
# Agregar trace
            fig.add_trace(go.Scatter(
                x=df_inst['TOTAL MATRCULA'],
                y=[area_map[area] for area in df_inst['REA CARRERA GENRICA']],
                mode='markers',
                name=inst,
                marker=marker_config,
                hovertemplate='%{customdata}<extra></extra>',
                customdata=hover_texts,
                showlegend=True
            ))
        
        # Configurar layout
        fig.update_layout(
            title=dict(
                text=f'Comparacin de Matrcula por rea - Ao {f_anio_g3.value}',
                font=dict(size=18, color='white'),
                x=0.5,
                xanchor='center'
            ),
            xaxis=dict(
                title=dict(text='Total Matrcula', font=dict(size=15, color='white')),
                tickfont=dict(size=11, color='lightgray'),
                gridcolor='#666666',
                gridwidth=0.5,
                griddash='dash',
                zeroline=False,
                tickformat=',',
                range=[0, df_plot['TOTAL MATRCULA'].max() * 1.25]
            ),
            yaxis=dict(
                title=dict(text='rea Carrera Genrica', font=dict(size=15, color='white')),
                tickfont=dict(size=13, color='white'),
                gridcolor='#444444',
                gridwidth=0.8,
                griddash='dash',
                zeroline=False,
                categoryorder='array',
                categoryarray=areas_formateadas
            ),
            plot_bgcolor='#1a1a1a',
            paper_bgcolor='#121212',
            hovermode='closest',
            legend=dict(
                title=dict(text='Instituciones', font=dict(size=12, color='white')),
                font=dict(size=11, color='white'),
                bgcolor='#2a2a2a',
                bordercolor='gray',
                borderwidth=1,
                x=1.01,
                y=1,
                xanchor='left',
                yanchor='top'
            ),
            height=max(600, len(areas_ordenadas) * 40),
            width=1800,
            margin=dict(l=280, r=180, t=80, b=80),
            autosize=False
        )
        
        # Reorganizar leyenda para que USS aparezca primero
        if uss_name in [trace.name for trace in fig.data]:
            traces_reordenadas = []
            for trace in fig.data:
                if trace.name == uss_name:
                    traces_reordenadas.insert(0, trace)
                else:
                    traces_reordenadas.append(trace)
            fig.data = traces_reordenadas
        
        fig.show()
    
    # Estadsticas
    with out_stats3:
        clear_output(wait=True)
        print("=" * 80)
        print(" ESTADSTICAS DEL GRFICO 3")
        print("=" * 80)
        print(f" Nivel 1: {list(f_niv1_g3.value)}")
        print(f" Nivel 2: {list(f_niv2_g3.value)}")
        print(f" Nivel 3: {list(f_niv3_g3.value)}")
        print(f" Ao: {f_anio_g3.value}")
        print(f" Regin: {list(f_region_g3.value)}")
        print(f" Nivel Global: {list(f_nivel_global_g3.value)}")
        print(f" Jornada: {list(f_jornada_g3.value)}")
        print(f" Instituciones comparadas: {instituciones_seleccionadas}")
        print(f"\n Total reas mostradas: {len(areas_ordenadas)}")
        print(f" Total Matrcula: {temp['TOTAL MATRCULA'].sum():,}")
        
        # Desglose por institucin
        print("\n Matrcula por Institucin:")
        for inst in instituciones_seleccionadas:
            mat_inst = temp[temp['NOMBRE INSTITUCIN'] == inst]['TOTAL MATRCULA'].sum()
            if mat_inst > 0:
                porcentaje = (mat_inst / temp['TOTAL MATRCULA'].sum()) * 100
                simbolo = "" if inst == uss_name else "  "
                print(f"{simbolo} {inst}: {mat_inst:,} ({porcentaje:.1f}%)")
        print("=" * 80)

btn_update3.on_click(actualizar_grafico3)

# ============================================
# Instrucciones y Dashboard
# ============================================
titulo3 = widgets.HTML(
    value="<h2 style='text-align:center; color:#2c3e50;'> Grfico 3: Comparacin por reas - USS vs Otras Instituciones</h2>"
)

instrucciones3 = widgets.HTML(value="""
<div style='background-color:#FFF3CD; padding:15px; border-radius:10px; border:2px solid #FFC107; margin:10px 0;'>
    <p style='margin:5px 0; font-size:14px;'><strong> Cmo usar este grfico:</strong></p>
    <ul style='margin:5px 0; font-size:13px;'>
        <li><strong>USS siempre est incluida</strong> (marcada en rojo con borde blanco)</li>
        <li>Selecciona <strong>mltiples instituciones</strong> para comparar (Ctrl/Cmd + clic)</li>
        <li>El eje Y muestra las <strong>reas de Carrera</strong></li>
        <li>El eje X muestra el <strong>Total de Matrcula</strong></li>
        <li>Cada color representa una institucin diferente</li>
    </ul>
</div>
""")

dashboard3 = VBox([
    titulo3,
    instrucciones3,
    accordion3,
    btn_update3,
    out_stats3,
    out_graph3
], layout=Layout(border='3px solid #28A745', padding='20px', border_radius='15px'))

display(dashboard3)

print(" Dashboard 3 listo. Compara USS con otras instituciones por rea.")
print(" Usa Ctrl/Cmd + clic para seleccionar mltiples instituciones.")
```
